## Exercise 11: Real estate price predictions

### Intermediate

```python
# input values for three mökkis:
#  - size [m^2],
#  - size of the sauna [m^2],
#  - distance to water [m],
#  - number of indoor bathrooms,
#  - proximity of neighbors [m]
X = [[66, 5, 15, 2, 500],
     [21, 3, 50, 1, 100],
     [120, 15, 5, 2, 1200]]

# coefficient values
c = [3000, 200 , -50, 5000, 100]

def predict(X, c):
    # write a loop that goes over the cabin data and for each cabin prints out
    # the predicted price of the cabin you can assume that the number of inputs
    # and the number of coefficients are the same
    for row in X:
        price = c[0]*row[0] + c[1]*row[1] + c[2]*row[2] + c[3]*row[3] + c[4]*row[4]
        print(price)


predict(X, c)
```

### Advanced

```python
# input values for three mökkis: size, size of sauna, distance to water, number of indoor bathrooms,
# proximity of neighbors
X = [[66, 5, 15, 2, 500],
     [21, 3, 50, 1, 100],
     [120, 15, 5, 2, 1200]]
c = [3000, 200, -50, 5000, 100]    # coefficient values

def predict(X, c):
    for row in X:
        price = sum(c[i] * row[i] for i in range(len(row)))
        print(price)

predict(X, c)
```

## Exercise 12: Least squares

### Intermediate

```python
import numpy as np

X = np.array([[66, 5, 15, 2, 500],
              [21, 3, 50, 1, 100],
              [120, 15, 5, 2, 1200]])
y = np.array([250000, 60000, 525000])
c = np.array([3000, 200 , -50, 5000, 100])    # coefficient values

def squared_error(X, y, c):
    sse = 0.0
    for xi, yi in zip(X, y):
        prediction = sum(xi * c)
        sse += (yi - prediction)**2

    print(sse)

squared_error(X, y, c)
```

### Advanced

```python
import numpy as np

# data
X = np.array([[66, 5, 15, 2, 500],
              [21, 3, 50, 1, 100],
              [120, 15, 5, 2, 1200]])
y = np.array([250000, 60000, 525000])

# alternative sets of coefficient values
c = np.array([[3000, 200 , -50, 5000, 100],
              [2000, -250, -100, 150, 250],
              [3000, -100, -150, 0, 150]])

def find_best(X, y, c):
    smallest_error = np.Inf
    best_index = -1
    for i, coeff in enumerate(c):
        sse = 0.0
        for xi, yi in zip(X, y):
            prediction = sum(xi * coeff)
            sse += (yi - prediction)**2

        if sse < smallest_error:
            smallest_error = sse
            best_index = i

    print("the best set is set %d" % best_index)


find_best(X, y, c)
```

## Exercise 13: Predictions with more data

### Intermediate

```python
import numpy as np


def main():
    np.set_printoptions(precision=1)

    x = np.array(
        [
            [25, 2, 50, 1, 500],
            [39, 3, 10, 1, 1000],
            [13, 2, 13, 1, 1000],
            [82, 5, 20, 2, 120],
            [130, 6, 10, 2, 600],
            [115, 6, 10, 1, 550]
        ]
    )

    y = np.array([127900, 222100, 143750, 268000, 460700, 407000])

    c = np.linalg.lstsq(x, y)[0]
    print(c)

    print(x @ c)

main()
```

### Advanced

```python
import numpy as np
from io import StringIO

input_string = '''
25 2 50 1 500 127900
39 3 10 1 1000 222100
13 2 13 1 1000 143750
82 5 20 2 120 268000
130 6 10 2 600 460700
115 6 10 1 550 407000
'''

np.set_printoptions(precision=1)    # this just changes the output settings for easier reading

def fit_model(input_file):
    data = np.genfromtxt(input_file, skip_header=1)

    x = data[:, :-1]   # all columns except the last: the features
    y = data[:, -1]    # the last column: the actual prices

    c, _, _, _ = np.linalg.lstsq(x, y, rcond=None)  # solve for coefficients

    print(c)
    print(x @ c)

# simulate reading a file
input_file = StringIO(input_string)
fit_model(input_file)
```

## Advanced

```python
import numpy as np
from io import StringIO


train_string = '''
25 2 50 1 500 127900
39 3 10 1 1000 222100
13 2 13 1 1000 143750
82 5 20 2 120 268000
130 6 10 2 600 460700
115 6 10 1 550 407000
'''

test_string = '''
36 3 15 1 850 196000
75 5 18 2 540 290000
'''

def main():
    np.set_printoptions(precision=1)    # this just changes the output settings for easier reading

    # read in the training data and separate it to x_train and y_train
    train_file = StringIO(train_string)
    train_data = np.genfromtxt(train_file, skip_header=1)
    x_train = train_data[:, :-1]
    y_train = train_data[:, -1]

    # fit a linear regression model to the data and get the coefficients
    c, _, _, _ = np.linalg.lstsq(x_train, y_train, rcond=None)

    # read in the test data and separate x_test from it
    test_file = StringIO(test_string)
    test_data = np.genfromtxt(test_file, skip_header=1)
    x_test = test_data[:, :-1]

    # print out the linear regression coefficients
    print(c)

    # this will print out the predicted prics for the two new cabins in the test data set
    print(x_test @ c)


main()
```
